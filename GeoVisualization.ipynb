# Folium + GeoJson 

```python
import folium
import pandas as pd
import geopandas as gpd
```

```python
#df = pd.read_json('australian_postcodes.json')
df = pd.read_csv('australian_postcodes.csv', dtype={'postcode': str})
display(df.head())
```

```python
# clean the post code for special purposes(PO Box)

df.drop(df[df['long']==0].index, inplace=True)
df.head()
```

```python
m = folium.Map([-38.043995, 145.264296], zoom_start=9)
m
```

```python
df_mel = df[df['postcode'].str[0]=='3']
display(df_mel.head())
for index, row in df_mel.iterrows():
    folium.Marker([row['lat'], row['long']],              
              tooltip=row['locality'],              
             ).add_to(m)
    
```

```python
df = pd.read_csv('Australian_Post_Codes_Lat_Lon.csv', dtype={'postcode': str})
display(df.head())
```

```python
#df[df['postcode']=='3084']
df['postcode'] = df['postcode'].str.zfill(4)
display(df.head())
```

```python
df_mel = df[df['postcode'].str[0]=='3']
display(df_mel.head())
for index, row in df_mel.iterrows():
    folium.Marker([row['lat'], row['lon']],              
              tooltip=row['suburb'],              
             ).add_to(m)
    
```
